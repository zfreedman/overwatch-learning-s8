## Import

```python
import codecs
from os import listdir
from os.path import isfile, join
```

## Files

```python
scraped_dirs = ["../data/hero_raw/", "../data/hero_processed/"]
scraped_file_lists = []
for i in range(len(scraped_dirs)):
    scraped_file_lists.append([(scraped_dirs[i] + f) for f in listdir(scraped_dirs[i]) if isfile(join(scraped_dirs[i], f))])

clean_dirs = ["../data/post_scrape_cleanup/raw/", "../data/post_scrape_cleanup/processed/"]
```

## Cleanup

```python
print(scraped_file_lists[1][0])
```

```python
tmp_str = "/some_thing/here/now/ana.csv"
tmp_str[tmp_str.rfind("/") + 1:]
```

```python
#Remove csv entries with improper length
#Correct length is assumed from length of first entry
#For each list of files
for i in range(len(scraped_file_lists)):
    file_list = scraped_file_lists[i]
    print("\n===\n===\nExploring {}\n\n".format(scraped_dirs[i]))
    #For each file
    for j in range(len(file_list)):
        dirty_file_str = file_list[j]
        clean_file_str = clean_dirs[i] + dirty_file_str[dirty_file_str.rfind("/") + 1:]
        data = []
        #Open file
        with codecs.open(dirty_file_str, "r", encoding="utf8") as f:
            #Get data
            data = f.readlines()
            #Count correct # of features
            feature_count = data[0].count(",")
            print("Analyzing {} entries from {}".format(len(data), dirty_file_str))
            print("Section: IMPROPER FEATURE COUNT")
            #Identify problem entries with incorrect feature count
            problem_indices = []
            for k in range(1, len(data)):
                if data[k].count(",") != feature_count:
                    problem_indices.append(k)
                    print(dirty_file_str, k+1, "{}/{}".format(data[k].count(","), feature_count))
            #Remove problem entries
            #Note: Loop backwards to avoid deleting wrong element...
            #Can do this with confidence because the data was looped
            #through originally from 0->end.
            print("Removing {} dirty entries.".format(len(problem_indices)))
            for k in range(len(problem_indices) - 1, -1, -1):
                del data[problem_indices[k]]
            
            #Identify problem entries which are duplicates (battletag-hero combos)
            print("Section: DUPLICATE ENTRIES")
            duplicate_indices = []
            btag_map = {}
            #For each entry
            for k in range(1, len(data)):
                #Create dictionary key for entry
                btag = data[k][:data[k].find(",")]
                #Test for duplicate
                if btag in btag_map:
                    duplicate_indices.append(k)
                    print("{}: {} duplicates {}.".format(btag, k+1, btag_map[btag]+1))
                else:
                    btag_map[btag] = k
            #Delete indices
            print("Removing {} duplicates.".format(len(duplicate_indices)))
            for k in range(len(duplicate_indices) - 1, -1, -1):
                del data[duplicate_indices[k]]
            
        #Write to cleaned file
        print("Writing remaining {} entries to {}.\n-------------".format(len(data), clean_file_str))
        with codecs.open(clean_file_str, "w", encoding="utf8") as f:
            f.writelines(data)
        #break
    #break
```
